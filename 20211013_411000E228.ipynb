# Python Conditions and If Statements

```python
a = 45
b = 256
if b < a:
    print("b is greater than a")
elif a < b:
    print("a is less than b")
```

```python
a = 78
b = 238
if b > a:
  print("b is greater than a")
```

```python
a = 67
b = 55
if b < a:
  print("b is less than a")
```

```python
a = 57
b = 57
if b < a:
    print("b is greater than a")
elif a < b:
    print("a is less than b")
else:
    print("a and b are equal")
```

#Leap Year

```python
year = eval(input("Please input the year"))

if ((year%400==0) or (year%4==0 and year%100!=0)):
    print("{0} is a leap year".format(year))
else:
  print("{0} is not a leap year".format(year))
```

#For Loop

```python
fruits = ["strawberry", "mango", "avocado"]
for x in fruits:
    print(x)
 
```

```python
for x in "avocado":
  print(x)
```

```python
for x in range(18):
    print(x)
```

```python
fruits = ["strawberry", "mango", "avocado"]
for x in fruits:
  print(x)
  if x == "mango":
    break
```

```python
for x in range(18):
  print(x)
  if x == 8:
    break
```

```python
for x in range(18):
  if x == 5:
    continue

  print(x)
```

```python
for x in range(18):
  if x%8 == 0:
    continue

  print(x)
```

```python
for x in range(10, 48, 37):
  print(x)
```

#While Loop
